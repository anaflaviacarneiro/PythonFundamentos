### Exercício 6

```python
# Considere os 3 arrays abaixo
# Retorne o valor do array xarr se o valor for True no array cond. Caso contrário, retorne o valor do array yarr.
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
ret = []
```

```python
for x , y, z  in zip(xarr, yarr, cond):
    if z == True:
        ret.append(x)
    else:
        ret.append(y)
ret
```

### Exercício 10

```python
# Crie um banco de dados no SQLite, crie uma tabela, insira registros, 
# consulte a tabela e retorne os dados em dataframe do Pandas
```

```python
#Cria o banco de dados
import os
os.remove("escola.db") if os.path.exists("escola.db") else None
```

```python
#conectando ao banco
import sqlite3

con = sqlite3.connect('escola.db')
```

```python
#Criando a tabela
cur = con.cursor()
sql_create = 'create table cursos '\
'(id integer primary key, '\
'titulo varchar(100), '\
'categoria varchar(140))'

cur.execute(sql_create)
```

```python
#inserindo valores 
# Dados
recset = [(100, 'Ciência de Dados', 'Data Science'),
          (101, 'Big Data Fundamentos', 'Big Data'),
          (102, 'Python Fundamentos', 'Análise de Dados')]
# Inserindo os registros
for rec in recset:
    cur.execute( 'insert into cursos values (?, ?, ?)', rec)
con.commit()
```

```python
import pandas as pd

dados = pd.DataFrame(cur.execute('select * from cursos'))
dados
```
